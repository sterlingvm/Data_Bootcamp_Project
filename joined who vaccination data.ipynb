```python
import pandas as pd
import numpy as np
```

```python
# load csv files
who_data = "who_data.csv"
vac_data = "vaccinations.csv"

# read files into DFs
who_df = pd.read_csv(who_data)
vac_df = pd.read_csv(vac_data)
```

```python
who_df.head()
```

```python
vac_df.head()
```

```python
who_df.rename(columns={'Name': 'COUNTRY'}, inplace=True)
who_df.head()
```

```python
# vlookup
joined_df = pd.merge(who_df, vac_df, on = 'COUNTRY', how='left')
joined_df.head()
```

```python
joined_df = joined_df.dropna()
joined_df.head()
```

```python
joined_df.to_csv("final_covid_data.csv", index=False)
```
