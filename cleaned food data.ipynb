```python
import pandas as pd
```

```python
food_data = "food_data.csv"
food_df = pd.read_csv(food_data)
```

```python
food_df.head()
```

```python
food_df.drop(["Animal fats","Aquatic Products, Other","Miscellaneous","Offals","Oilcrops","Pulses","Spices","Starchy Roots","Stimulants","Sugar Crops","Confirmed","Deaths","Recovered","Active","Population","Unit (all except Population)"], axis=1, inplace=True)
food_df
```

```python
food_df.rename(columns={'Country': 'COUNTRY'}, inplace=True)
food_df.head(10)
```

```python
# replace "<2.5" values in Undernourished column with zeroes 
food_df.loc[food_df.Undernourished=="<2.5", 'Undernourished'] = 0
food_df.head(10)
```

```python
# drop NA rows from food df
final_food_df = food_df.dropna()
final_food_df.head(20)
```

```python
final_food_df.to_csv("final_food_data.csv", index=False)
```
